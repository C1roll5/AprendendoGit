Importando as bibliotecas para o sistema
- pandas

```python
import pandas as pd
```

Com esse notebook vamos
- carregar as bases de dados
- limpar os dados
- mesclagem de dados(merge)
- agrupamento de dados (groupby)
- tabela dinamica

```python
df_sales_data = pd.read_csv('/content/sales_data.csv')
print(df_sales_data)
```

Base de dados "pproduct_inventory" excel

```python
df_product_inventory = pd.read_excel('/content/product_inventory.xlsx')
display(df_product_inventory)
```

completar dados de "price" ausentes com mdia do campo

```python
# calculo de mdia
media_precos = df_sales_data['Price'].mean()

# colocando o valor dentro da tabela
df_sales_data['Price'] = df_sales_data['Price'].fillna(media_precos)

# exibir os dados
display(df_sales_data)
```

relatorio com vendas e produtos

```python
#RENOMEARIA A COLUNA DO DF_PRODUCT
df_product_inventory.rename(columns={'ProductID':'Product_ID'}, inplace=True)

relatorio_final = pd.merge(df_sales_data,df_product_inventory, on='Product_ID')

# criar um novo campo - valor da venda
relatorio_final['Total_da_venda'] = relatorio_final['Quantity'] * relatorio_final['Price']

display(relatorio_final)

#salvando em excel
relatorio_final.to_csv('relatorio_final.csv', index=False)
```
